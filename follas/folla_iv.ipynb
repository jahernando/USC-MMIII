$r(t) = (2 \cos t,  2 \sin t, t); \; t0 = 2\pi$

```python
def line3d(ts, xfun, yfun, zfun, name = ''):
    xs = xfun(ts)
    ys = yfun(ts)
    zs = zfun(ts)
    fig = plt.figure(); ax = fig.gca(projection='3d')
    ax.plot(xs, ys, zs)
    plt.xlabel('$x$'); plt.ylabel('$y$'); ax.set_zlabel('z'); plt.axis('equal')
    if name != '': plt.title(name)
    plt.tight_layout();
    return
```

```python
ts = np.linspace(0, 8*np.pi, 100)
xf = lambda t : 2 * np.cos(t)
yf = lambda t : 2 * np.sin(t)
zf = lambda t : t
line3d(ts, xf, yf, zf)
```

**ejercicio 5**

$ r(u, v) = (u, v, v/2)$

$r(u, v) =(2 u \cos v,  2u \sin v, \frac{u^2}{2})$

$r(u, v) = (2 \cos u, v, 2 \sin u)$

$r(u, v) = (3 \cos v \cos u, 3 \cos v \sin u, 5 \sin v)$

```python
def surface(us, vs, xfun, yfun, zfun):
    ums, vms = np.meshgrid(us, vs)
    xs = xfun(ums, vms)
    ys = yfun(ums, vms)
    zs = zfun(ums, vms)
    fig = plt.figure(); ax = fig.gca(projection='3d')
    ax.plot_wireframe(xs, ys, zs, color = 'blue', alpha = 0.8)
    plt.xlabel('$x$'); plt.ylabel('$y$'); ax.set_zlabel('z'); ax.set_aspect('equal')
    return
```

```python
us = np.linspace(0., 2., 20)
vs = np.linspace(0., 2.*np.pi, 20)
xf = lambda u, v : 2 * u * np.cos(v)
yf = lambda u, v : 2 * u * np.sin(v)
zf = lambda u, v : u**2/2 + v*0
surface(us, vs, xf, yf, zf)
```

```python
us = np.linspace(0., 2.*np.pi, 20)
vs = np.linspace(-1,       1., 20)
xf = lambda u, v : 2 * np.cos(u)
yf = lambda u, v : v + 0 *u
zf = lambda u, v : 2 * np.sin(u)
surface(us, vs, xf, yf, zf)
```

```python
us = np.linspace(0.      , 2.*np.pi, 20)
vs = np.linspace(-np.pi/2,  np.pi/2, 20)
xf = lambda u, v : 3 * np.cos(v) * np.cos(u)
yf = lambda u, v : 3 * np.cos(v) * np.sin(u)
zf = lambda u, v : 5 * np.sin(v)
surface(us, vs, xf, yf, zf)
```

```python
us = np.linspace(-2., 2., 20)
vs = np.linspace(-2., 2., 20)
xf = lambda u, v : u + 0 * v
yf = lambda u, v : v + 0 * u
zf = lambda u, v : v/2 + 0 * u
surface(us, vs, xf, yf, zf)
```

```python
c = 0.5
r = np.sqrt(c)
rs   = np.linspace(0, r, 100)
phis = np.linspace(0, 2 * np.pi, 100)
rms, phims = np.meshgrid(rs, phis)
xms   = rms * np.cos(phims)
yms   = rms * np.sin(phims)
zms   = np.sqrt(r * r - rms * rms )
ax    = plt.gca(projection='3d')
ax.plot_wireframe(xms, yms,  zms, alpha = 0.5, color = 'b')
ax.plot_wireframe(xms, yms, -zms, alpha = 0.5, color = 'b')
ax.set_aspect('equal');
```

```python
c = 1.
r = np.sqrt(c)
rs   = np.linspace(r, 2*r, 100)
phis = np.linspace(0, 2 * np.pi, 100)
rms, phims = np.meshgrid(rs, phis)
xms   = rms * np.cos(phims)
yms   = rms * np.sin(phims)
zms   = np.sqrt(- r * r + rms * rms )
ax    = plt.gca(projection='3d')
ax.plot_wireframe(xms, yms,  zms, alpha = 0.5, color = 'b')
ax.plot_wireframe(xms, yms, -zms, alpha = 0.5, color = 'b')
ax.set_aspect('equal');
```
