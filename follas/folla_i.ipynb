### 1.6 

$f(x, y) = \max(|x|, |y|)$

Los conjuntos de nivel son cuadrados centrados en el origen con vÃ©rtices en (c, c), (-c, c), (-c, -c), (c, -c)

```python
def fun(x, y):
    return np.maximum(np.abs(x), np.abs(y))

xrange, yrange = (-10., 10., 100), (-10., 10., 100)
#gf.graph   (fun, xrange, yrange);

xs = np.linspace(-10., 10., 100)
ys = np.linspace(-10., 10., 100)
xms, yms = np.meshgrid(xs, ys)
zms = fun(xms, yms)

ax = plt.gca(projection='3d')
ax.plot_surface(xms, yms, zms);
plt.figure()
plt.contour(xs, ys, zms);
plt.gca().set_aspect('equal');
#plt.aspect('equal')
```

### 1.9

$f(x,y) = \sqrt{\frac{1}{y} - x^2}$

Se tiene que cumplir que $\frac{1}{y} - x^2 \ge 0$, esto obliga a  $y \le \frac{1}{x^2}$ e $y > 0$

$f(x,y) = \sqrt{\frac{1}{y} - x^2} = c, \;\; \frac{1}{y} - x^2 = c^2$

$y = \frac{1}{c^2+x^2}$


```python
def fun(x, y):
    val = np.maximum(1/y - x*x, 0.)
    val = np.sqrt(val)
    return val
xrange, yrange = (-5., 5., 100), (0.1, 5., 100)
gf.graph  (fun, xrange, yrange);
gf.contour(fun, xrange, yrange);
```

### 2.3

$f(x, y, z) = x^2 + y^2 + z^2 = c$

  * $c = 0$, un solo punto, el origen, $(0, 0, 0)$
  
  * $c = 1$, una esfera de radio unidad 

```python
c = 0.5
r = np.sqrt(c)
rs   = np.linspace(0, r, 100)
phis = np.linspace(0, 2 * np.pi, 100)
rms, phims = np.meshgrid(rs, phis)
xms   = rms * np.cos(phims)
yms   = rms * np.sin(phims)
zms   = np.sqrt(r * r - rms * rms )
ax    = plt.gca(projection='3d')
ax.plot_wireframe(xms, yms,  zms, alpha = 0.5, color = 'b')
ax.plot_wireframe(xms, yms, -zms, alpha = 0.5, color = 'b')
#ax.set_aspect('equal');
```

### 2.4

$f(x, y, z) = x^2 + y^2 - z^2 = c$

  * $c = 0$, se trata de un cÃ³nica, $z = \sqrt{x^2 + y^2}$, con conexiÃ³n en el origen
  
  * $c = 1$, se trata de un paraboliode, con conexiÃ³n en el $z=0$ con una circunferencia de radio unidad.  
  $z^2 = x^2 + y^2 - c$, caso $z = 0, \;\; x^2 + y^2 = c$
  
  * $c = -1$, se trata de un paraboloide con dos hojas desconectadas, la superior y la inferior.
  $x^2 + y^2 = z^2 - |c|$, para $|c| = 1, c = -1$, tendrÃ­amos dos puntos $(0, 0, \pm 1)$. Y en $z = \pm 2$, dos circunferencias en $(x, y)$ de radio unidad.

```python
c = 1.
r = np.sqrt(c)
rs   = np.linspace(r, 2*r, 100)
phis = np.linspace(0, 2 * np.pi, 100)
rms, phims = np.meshgrid(rs, phis)
xms   = rms * np.cos(phims)
yms   = rms * np.sin(phims)
zms   = np.sqrt(- r * r + rms * rms )
ax    = plt.gca(projection='3d')
ax.plot_wireframe(xms, yms,  zms, alpha = 0.5, color = 'b')
ax.plot_wireframe(xms, yms, -zms, alpha = 0.5, color = 'b')
#ax.set_aspect('equal');
```

### 2.5

$f(x, y, z) = x^2 + 4 y^2 + 3z^2$

se trata de un elipsoide

 * c = 0, de un solo punto (0, 0, 0)

 * c = 1, es un elipsoide, $x^2 + 4 y^2 + 3z^2 = 1$
 
 Notar que si cambiamos las bariables $y' = 2 y, \; z' = \sqrt{3} z$, queda: $x^2 + y'^2 + z'^2 = 1$, una esfera de radio unidad. Luego se trata de coger una esfera y de achatar en este caso sus ejes: $y = y'/2, \; z = z'/\sqrt{3}$.

```python
cc = 2.
a, b, c = np.sqrt(cc), np.sqrt(cc/4.), np.sqrt(cc/3.)
a, b, c = 1., 1., 0.1
rs   = np.linspace(0, 1., 40)
phis = np.linspace(0, 2 * np.pi, 40)
rms, phims = np.meshgrid(rs, phis)
xms   = rms * np.cos(phims)/a
yms   = rms * np.sin(phims)/b
zms   = np.sqrt(1 - rms * rms)/c
ax    = plt.gca(projection='3d')
ax.plot_wireframe(xms, yms,  zms, alpha = 0.5, color = 'b')
ax.plot_wireframe(xms, yms, -zms, alpha = 0.5, color = 'b')
```
